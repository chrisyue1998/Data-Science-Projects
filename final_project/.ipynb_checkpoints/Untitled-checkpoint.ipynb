```python
import pandas as pd
import numpy as np
```

```python
table = pd.read_csv('winemag-data_first150k.csv')
table.drop(['Unnamed: 0'], axis=1, inplace=True)
table = table[['country', 'description', 'province', 'region_1', 'region_2', 'winery', 
               'designation', 'price', 'points', 'variety']]
table = table.dropna()
table
```

```python
varieties = {}
```

```python
def discretize(col_name):
    conversion = {}
    cur_num = 0
    for index, row in table.iterrows():
        if row[col_name] not in conversion.keys():
            if col_name == 'variety':
                varieties[cur_num] = row[col_name]
                
            conversion[row[col_name]] = cur_num
            cur_num += 1
        table.at[index, col_name] = conversion[row[col_name]]
```

```python
to_discretize = ['country', 'province', 'region_1', 'region_2', 'winery', 'designation', 'variety']

for col_name in to_discretize:
    discretize(col_name)
    
table.head()
```
