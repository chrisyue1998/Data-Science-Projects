### Problem 1

```python
%config IPCompleter.greedy=True
%matplotlib inline

import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
sqlite_file = 'lahman2014.sqlite'
conn = sqlite3.connect(sqlite_file)

query = "SELECT yearID, teamID, sum(salary) as total_payroll FROM Salaries GROUP BY teamID, yearID"
stats = pd.read_sql(query, conn)

query = "SELECT yearID, teamID, W, G, (100 * W / G) as win_percentage FROM Teams GROUP BY teamID, yearID"

# inner join to remove any entries with nan values
stats = stats.merge(pd.read_sql(query, conn), how='inner', on=['yearID', 'teamID'])

# reorder and rename DataFrame columns
stats = stats[['yearID', 'teamID', 'W', 'G', 'win_percentage', 'total_payroll']]
stats.columns = ['year', 'team', 'W', 'G', 'win_percentage', 'total_payroll']

# limit entries in DataFrame to only ones that are in the 1990 to 2014 timeframe
stats = stats.loc[(stats['year'] >= 1990) & (stats['year'] <= 2014)]
stats
```

This code retrieves two Pandas DataFrames containing the years, teams, and total payrolls of each team from the Salaries table and the years, teams, wins, total games played, and win percentages of each team from the Teams table, groups the results by the team names and the years, and performs an inner join on the two results to combine the DataFrames and eliminate any missing values.

```python
# get a list of all teams in time frame
teams = pd.DataFrame(np.unique(stats['team'].values))
teams.columns = ['team']
teams.index = range(0, len(teams))

fig, axs = plt.subplots(len(teams), figsize=(5, 90), constrained_layout=True)

# plot each team's total payrolls throughout the time frame
for index, row in teams.iterrows():
    team_plot = stats.loc[stats['team'] == teams['team'][index]]
    team_plot = team_plot[['year', 'total_payroll']]
    plot = sns.barplot(x=team_plot['year'], y=team_plot['total_payroll'], ax=axs[index], 
                palette=sns.color_palette('muted', 1))
    axs[index].set_title(teams['team'][index])
    
    for item in plot.get_xticklabels():
        item.set_rotation(60)
```

This code iterates through a list of all of the teams and plots the total payrolls of each team from 1990 to 2014.

### Problem 4

```python
# label each data point with the corresponding team name
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))
```

```python
fig, axs = plt.subplots(nrows=5, figsize=(18, 40), constrained_layout=True)
ax_idx = 0

# iterate over and plot 5-year intervals
for i in range(1990, 2015, 5):
    payroll_win_percentage_arr = [[], []]
    cur_interval_stats = stats.loc[(stats['year'] >= i) & (stats['year'] < i + 5)]
    existing_teams = cur_interval_stats['team'].drop_duplicates()
    
    for index, row in teams.iterrows():
        cur_mean_payroll = cur_interval_stats['total_payroll'].loc[cur_interval_stats['team'] == row['team']].mean()
        cur_mean_win_percentage = cur_interval_stats['win_percentage'] \
        .loc[cur_interval_stats['team'] == row['team']].mean()
        
        # skip nans to ensure that no nan values raise an error when polyfit is called
        if not math.isnan(cur_mean_payroll) and not math.isnan(cur_mean_win_percentage):
            payroll_win_percentage_arr[0].append(cur_mean_payroll)
            payroll_win_percentage_arr[1].append(cur_mean_win_percentage)
            
    # cast to an NumPy array to transpose and store data in DataFrame
    payroll_win_percentage_arr = np.array(payroll_win_percentage_arr)
    
    payroll_win_percentage_plot = pd.DataFrame(np.transpose(payroll_win_percentage_arr))
    payroll_win_percentage_plot.columns = ['mean_total_payroll', 'mean_win_percentage']
    x = payroll_win_percentage_plot['mean_total_payroll']
    y = payroll_win_percentage_plot['mean_win_percentage']
    
    # calculate regression line
    trend = np.polyfit(x=x, y=y, deg=1)
    trend = trend[1] + trend[0] * x
    trend_plot = pd.DataFrame(np.transpose([[x_val for x_val in x], [trend_val for trend_val in trend]]))
    trend_plot.columns = ['mean_total_payroll', 'mean_win_percentage']
    
    # plot data and regression line
    ax = sns.scatterplot(x=payroll_win_percentage_plot['mean_total_payroll'], y=payroll_win_percentage_plot
                         ['mean_win_percentage'], ax=axs[ax_idx])
    sns.lineplot(x=trend_plot['mean_total_payroll'], y=trend_plot['mean_win_percentage'], ax=ax, color='r')
    
    X = payroll_win_percentage_plot['mean_total_payroll']
    Y = payroll_win_percentage_plot['mean_win_percentage']
    names = existing_teams
    ymin, ymax = ax.get_ylim()
    color="#3498db"
    bonus = (ymax - ymin) / 20 # displacement
    for x, y, name in zip(X, Y, names):
        ax.text(x, y + bonus, name, color=color, rotation=60)

    
    if i + 5 == 2015:
        ax.set_title(str(i) + '-' + str(i + 4))
    else:
        ax.set_title(str(i) + '-' + str(i + 5))
        
    ax_idx += 1
```

This code calculates and plots the mean win percentage for each team versus the mean total payroll for each team in five year intervals from 1990 to 2014. The code also labels each data point with the corresponding team name.

## Part 3: Data Transformations

### Problem 5

```python
# same code as 2 cells above, but standardizes teams' payrolls

fig, axs = plt.subplots(nrows=5, figsize=(18, 45), constrained_layout=True)
ax_idx = 0

stats['standardized_payroll'] = np.nan

# map each year to a tuple containing the average payroll and standard deviation of payrolls for that year
years_avg_std = {}

for year in range(1990, 2015):
    year_stats = stats.loc[stats['year'] == year]
    year_payroll_avg = year_stats['total_payroll'].mean()
    year_payroll_std = year_stats['total_payroll'].std()
    years_avg_std[year] = (year_payroll_avg, year_payroll_std)
    
for index, row in stats.iterrows():
    avg = years_avg_std[row['year']][0]
    std = years_avg_std[row['year']][1]
    standardized = (row['total_payroll'] - avg) / std
    stats.set_value(index, 'standardized_payroll', standardized)

# iterate over and plot 5-year intervals
for i in range(1990, 2015, 5):
    payroll_win_percentage_arr = [[], []]
    cur_interval_stats = stats.loc[(stats['year'] >= i) & (stats['year'] < i + 5)]
    existing_teams = cur_interval_stats['team'].drop_duplicates()
    
    
    for index, row in teams.iterrows(): 
        cur_std_payroll = cur_interval_stats['standardized_payroll'] \
            .loc[cur_interval_stats['team'] == row['team']].mean()
        cur_mean_win_percentage = cur_interval_stats['win_percentage'] \
            .loc[cur_interval_stats['team'] == row['team']].mean()
        
        # skips nans to ensure that no nan values raise an error when polyfit is called
        if not math.isnan(cur_std_payroll) and not math.isnan(cur_mean_win_percentage):
            payroll_win_percentage_arr[0].append(cur_std_payroll)
            payroll_win_percentage_arr[1].append(cur_mean_win_percentage)
            
    payroll_win_percentage_arr = np.array(payroll_win_percentage_arr)
    
    payroll_win_percentage_plot = pd.DataFrame(np.transpose(payroll_win_percentage_arr))
    payroll_win_percentage_plot.columns = ['mean_standardized_payroll', 'mean_win_percentage']
    x = payroll_win_percentage_plot['mean_standardized_payroll']
    y = payroll_win_percentage_plot['mean_win_percentage']
    
    trend = np.polyfit(x=x, y=y, deg=1)
    trend = trend[1] + trend[0] * x
    trend_plot = pd.DataFrame(np.transpose([[x_val for x_val in x], [trend_val for trend_val in trend]]))
    trend_plot.columns = ['mean_standardized_payroll', 'mean_win_percentage']
    
    ax = sns.scatterplot(x=payroll_win_percentage_plot['mean_standardized_payroll'], y=payroll_win_percentage_plot
                         ['mean_win_percentage'], ax=axs[ax_idx])
    sns.lineplot(x=trend_plot['mean_standardized_payroll'], y=trend_plot['mean_win_percentage'], ax=ax, color='r')
    
    X = payroll_win_percentage_plot['mean_standardized_payroll']
    Y = payroll_win_percentage_plot['mean_win_percentage']
    names = existing_teams
    ymin, ymax = ax.get_ylim()
    color="#3498db"
    bonus = (ymax - ymin) / 20 # displacement
    for x, y, name in zip(X, Y, names):
        ax.text(x, y + bonus, name, color=color, rotation=60)

    
    if i + 5 == 2015:
        ax.set_title(str(i) + '-' + str(i + 4))
    else:
        ax.set_title(str(i) + '-' + str(i + 5))
        
    ax_idx += 1
```

This code performs the same steps for plotting as the code for Problem 4 does, but it standardizes the payrolls for each team, for each year, stores the standardized payroll in a new columns in the stats DataFrame, and plots the mean win percentage versus the mean standardized payroll for each team instead of the mean payroll.

The plots in problem 6 allow one to more accurately analyze each team's spending habits; rather than simply display how much a team is spending on average in a given time period such as in problem 4, the new plots show how much each team is spending in a time period relative to the mean of all of the payrolls in that time period. ### Problem 8

```python
# creates a new "efficiency" column in the stats DataFrame
stats['efficiency'] = np.nan

for index, row in stats.iterrows():
    win_pct = row['win_percentage']
    std_payroll = row['standardized_payroll']
    expected_win_pct = 50 + 2.5 * std_payroll
    stats.set_value(index, 'efficiency', expected_win_pct)

teams = ['OAK', 'BOS', 'NYA', 'ATL', 'TBA']

fig, ax = plt.subplots(figsize=(10, 6))

# plot the efficiency for each team throughout the Moneyball period
for team in teams:
    cur_team_stats = stats.loc[stats['team'] == team]
    efficiency_plot = np.array([cur_team_stats['year'], cur_team_stats['efficiency']])
    efficiency_plot = pd.DataFrame(np.transpose(efficiency_plot))
    efficiency_plot.columns = ['year', 'efficiency']
    sns.lineplot(x=efficiency_plot['year'], y=efficiency_plot['efficiency'], ax=ax, label=team)
    ax.legend()
```

This code creates a new column representing the spending efficiency of each team, for each year, and uses the formula for expected win percentage as derived from the regression line in Problem 7 to calculate the spending efficiency of each team. It then plots the spending efficiency of OAK, BOS, NYA, ATL, and TBA between the years 1990 and 2014.
